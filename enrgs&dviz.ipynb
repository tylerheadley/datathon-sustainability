```python
import pandas as pd
import csv

# Load your dataset
try:
    df = pd.read_csv("train_features.csv")
except pd.errors.ParserError as e:
    print(f"Error encountered: {e}")

    # Find the problematic line in the file
    line_number = int(str(e).split('line ')[1].split(',')[0])  # Extract line number
    with open("train_features.csv", 'r') as file:
        for i, line in enumerate(file):
            if i + 1 == line_number:
                print(f"Problematic line {line_number}: {line}")
                reader = csv.reader([line])
                try:
                    num_fields = len(next(reader))
                except csv.Error as e2:
                    print(f"Error parsing the line: {e2}")
                else:
                    print(f"Number of fields in the problematic line: {num_fields}")
                break

# Identify historical weather columns (they contain '_h' in their name)
historical_weather_cols = ["is_consumption","datetime","county","target"]

# Keep only rows where at least one historical weather column is not null
df_historical_only = df[df["is_consumption"].notna()]

# Optionally keep datetime, county, data_block_id for reference
reference_cols = ['datetime', 'county', 'data_block_id']
reference_cols = [col for col in reference_cols if col in df.columns]
df_historical_only = df_historical_only[reference_cols + historical_weather_cols]

# Sanity check on time zone if needed
# df["datetime"] = pd.to_datetime(df["datetime"], utc=True)  # only if necessary

# Convert 'target' column to numeric, handling errors
df['target'] = pd.to_numeric(df['target'], errors='coerce')

# Step 1: Aggregate total consumption and production per hour
agg_df = df.groupby(["datetime", "is_consumption"])["target"].sum().reset_index()

# Convert 'datetime' to datetime objects before pivoting, ensuring consistent timezone
agg_df['datetime'] = pd.to_datetime(agg_df['datetime'], errors='coerce', utc=True)
# errors='coerce' will set invalid parsing to NaT (Not a Time)
# utc=True makes all datetimes timezone-aware with UTC


# *** Before pivoting, check for and handle duplicate datetime/is_consumption combinations ***
# Here, we choose to sum the 'target' for duplicate combinations
agg_df = agg_df.groupby(["datetime", "is_consumption"])["target"].sum().reset_index()

# Step 2: Pivot the table so you have two columns: consumption and production
pivot_df = agg_df.pivot(index="datetime", columns="is_consumption", values="target").rename(
    columns={True: "consumption", False: "production"}
)

# Optional: Fill missing values with 0 (or NaN depending on your needs)
pivot_df = pivot_df.fillna(0)

# Step 3: Check for missing hourly timestamps
# First, create a complete datetime range from min to max in hourly steps
# *** Filter out NaT values before calculating min/max ***
valid_index = pivot_df.index[pivot_df.index.notna()]
full_index = pd.date_range(start=valid_index.min(), end=valid_index.max(), freq="H")

# Compare this full index to what exists in the data
missing_timestamps = full_index.difference(pivot_df.index)

# Output results
print(f"Total expected hourly timestamps: {len(full_index)}")
print(f"Actual timestamps in data: {pivot_df.shape[0]}")
print(f"Missing timestamps: {len(missing_timestamps)}")

# Optional: Save the result for future use
pivot_df.to_csv("estonia_aggregated_by_hour.csv")
```

# Weather Forecast Accuracy and Energy Imbalance

Below is an example Python script that uses pandas to load your CSV files, computes a basic forecast accuracy metric based on temperature (by comparing the forecasted temperature with the historical temperature), merges the data with electricity prices, and then produces a scatter plot to visualize the correlation between forecast accuracy (measured as the absolute error in temperature) and electricity prices. You can adapt the metric (or add others) as needed.

## Does it correlate with temperature forecast? 0.15

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# Set plot style and parameters
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (19, 11)
plt.rcParams["font.size"] = 20
plt.rcParams["xtick.color"] = 'black'
plt.rcParams["ytick.color"] = 'black'
plt.rcParams["axes.edgecolor"] = 'black'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams["axes.labelcolor"] = 'black'
plt.rcParams["axes.linewidth"] = 1

# -------------------------------
# Step 1: Load the combined CSV file
# -------------------------------
df = pd.read_csv("train_features.csv", parse_dates=['datetime', 'forecast_date_electricity', 'origin_date_electricity'])

# -------------------------------
# Step 2: Compute forecast accuracy metric
# -------------------------------
# Calculate temperature forecast relative error (absolute difference between forecast and historical temperature).
df["temp_error"] = ((df["temperature_f_mean"] - df["temperature_h_mean"])).abs()

# Optionally, create a binary flag using a threshold (e.g., 1°C).
accuracy_threshold = 1.0
df["accurate"] = df["temp_error"] < accuracy_threshold

# -------------------------------
# Step 3: Compute Energy Imbalance by Aggregating per Datetime
# -------------------------------
# Here we assume:
#   - Rows with is_consumption == 0 (or False) indicate production.
#   - Rows with is_consumption == 1 (or True) indicate consumption.
# The energy imbalance is defined as: Production - Consumption.
#
# We also compute the average temperature forecast error per datetime.

df_grouped = df.groupby("datetime").apply(lambda group: pd.Series({
    "avg_temp_error": group["temp_error"].mean(),
    "abs_imbalance": abs(group.loc[group["is_consumption"] == 0, "target"].sum() -
                 group.loc[group["is_consumption"] == 1, "target"].sum())
})).reset_index()

# -------------------------------
# Step 4: Filter Out Outliers in Imbalance
# -------------------------------
# Here we remove extreme imbalance values by filtering rows between the 1st and 99th percentiles.
imbalance_lower = df_grouped["abs_imbalance"].quantile(0.01)
imbalance_upper = df_grouped["abs_imbalance"].quantile(0.99)
df_filtered = df_grouped[(df_grouped["abs_imbalance"] >= imbalance_lower) & (df_grouped["abs_imbalance"] <= imbalance_upper)]

# -------------------------------
# Step 5: Graph the correlation using a hexbin plot
# -------------------------------
plt.figure(figsize=(8, 6))
hb = plt.hexbin(df_filtered["avg_temp_error"], df_filtered["abs_imbalance"],
                gridsize=30, cmap='viridis', norm=LogNorm())
plt.xlabel("Average Temperature Forecast Error (°C)")
plt.ylabel("Energy Imbalance")
plt.title("Hexbin Plot: Temperature Forecast Error vs Abs Energy Imbalance")
cb = plt.colorbar(hb)
cb.set_label("Count/Bin (log)")
plt.grid(True)
plt.show()

# -------------------------------
# Step 6: Compute the correlation coefficient
# -------------------------------
correlation = df_filtered["avg_temp_error"].corr(df_filtered["abs_imbalance"])
print("Correlation coefficient between average temperature forecast error and energy imbalance:", correlation)
```

# What about Combined Forecast Error? 0.01

Below is the adapted code that computes a “combined_error” as the sum of all forecast errors for various weather parameters. In this example, we sum the absolute errors for temperature, dewpoint, cloudcover (high, low, mid, and total), direct solar radiation, and snowfall. (You can further extend this by adding other forecast errors if appropriate.) The script then computes the average combined error and absolute energy imbalance per datetime, filters out extreme values, and visualizes the relationship using a hexbin plot with logarithmic normalization.

- Use relative error so you can sum the values which have different scales.

- Only plot if there is both a forecasted and historical value.
- Remove extreme imbalance values by keeping rows between the 1st and 99th percentiles.
- Computing correlation .corr properly normalizes ✅

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# Set plot style and parameters
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (19, 11)
plt.rcParams["font.size"] = 20
plt.rcParams["xtick.color"] = 'black'
plt.rcParams["ytick.color"] = 'black'
plt.rcParams["axes.edgecolor"] = 'black'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams["axes.labelcolor"] = 'black'
plt.rcParams["axes.linewidth"] = 1

# -------------------------------
# Step 1: Load the combined CSV file
# -------------------------------
df = pd.read_csv("train_features.csv",
                 parse_dates=['datetime', 'forecast_date_electricity', 'origin_date_electricity'])

# -------------------------------
# Step 2: Filter for Valid Forecast-Historical Pairs
# -------------------------------
# Define the parameters for which we want to compute relative errors.
params = ["temperature", "dewpoint", "cloudcover_high", "cloudcover_low",
          "cloudcover_mid", "cloudcover_total", "direct_solar_radiation", "snowfall"]

# Construct forecast and historical column names.
forecast_cols = [f"{p}_f_mean" for p in params]
historical_cols = [f"{p}_h_mean" for p in params]

# Only keep rows that have non-null values in all forecast and corresponding historical columns.
mask = df[forecast_cols + historical_cols].notna().all(axis=1)
df = df[mask]

# -------------------------------
# Step 3: Compute Combined Relative Forecast Error
# -------------------------------
epsilon = 1e-5  # Small constant to avoid division by zero

# Compute relative (normalized) errors for each parameter.
temp_rel_error = (df["temperature_f_mean"] - df["temperature_h_mean"]).abs() / (df["temperature_h_mean"].abs() + epsilon)
dewpoint_rel_error = (df["dewpoint_f_mean"] - df["dewpoint_h_mean"]).abs() / (df["dewpoint_h_mean"].abs() + epsilon)
cloudcover_high_rel_error = (df["cloudcover_high_f_mean"] - df["cloudcover_high_h_mean"]).abs() / (df["cloudcover_high_h_mean"].abs() + epsilon)
cloudcover_low_rel_error = (df["cloudcover_low_f_mean"] - df["cloudcover_low_h_mean"]).abs() / (df["cloudcover_low_h_mean"].abs() + epsilon)
cloudcover_mid_rel_error = (df["cloudcover_mid_f_mean"] - df["cloudcover_mid_h_mean"]).abs() / (df["cloudcover_mid_h_mean"].abs() + epsilon)
cloudcover_total_rel_error = (df["cloudcover_total_f_mean"] - df["cloudcover_total_h_mean"]).abs() / (df["cloudcover_total_h_mean"].abs() + epsilon)
direct_solar_radiation_rel_error = (df["direct_solar_radiation_f_mean"] - df["direct_solar_radiation_h_mean"]).abs() / (df["direct_solar_radiation_h_mean"].abs() + epsilon)
snowfall_rel_error = (df["snowfall_f_mean"] - df["snowfall_h_mean"]).abs() / (df["snowfall_h_mean"].abs() + epsilon)

# Sum some relative errors to get a combined error.
df["combined_error"] = (temp_rel_error + cloudcover_total_rel_error +
                        direct_solar_radiation_rel_error + snowfall_rel_error)

# Optionally, create a binary flag for "accurate" forecasts based on a threshold.
accuracy_threshold = 1.0  # adjust as needed
df["accurate"] = df["combined_error"] < accuracy_threshold

# -------------------------------
# Step 4: Compute Energy Imbalance by Aggregating per Datetime
# -------------------------------
# We assume:
#   - Rows with is_consumption == 0 indicate production.
#   - Rows with is_consumption == 1 indicate consumption.
# Energy imbalance is defined as: Production - Consumption.
# For each datetime, compute:
#   - The average combined relative error.
#   - The absolute energy imbalance.
df_grouped = df.groupby("datetime").apply(lambda group: pd.Series({
    "avg_combined_error": group["combined_error"].mean(),
    "abs_imbalance": abs(
        group.loc[group["is_consumption"] == 0, "target"].sum() -
        group.loc[group["is_consumption"] == 1, "target"].sum()
    )
})).reset_index()

# -------------------------------
# Step 5: Filter Out Outliers in Imbalance
# -------------------------------
# Remove extreme imbalance values by keeping rows between the 1st and 99th percentiles.
imbalance_lower = df_grouped["abs_imbalance"].quantile(0.01)
imbalance_upper = df_grouped["abs_imbalance"].quantile(0.99)
df_filtered = df_grouped[(df_grouped["abs_imbalance"] >= imbalance_lower) &
                         (df_grouped["abs_imbalance"] <= imbalance_upper)]

# -------------------------------
# Step 6: Graph the Correlation Using a Hexbin Plot
# -------------------------------
plt.figure(figsize=(8, 6))
hb = plt.hexbin(df_filtered["avg_combined_error"], df_filtered["abs_imbalance"],
                gridsize=30, cmap='viridis', norm=LogNorm())
plt.xlabel("Average Combined Relative Forecast Error")
plt.ylabel("Absolute Energy Imbalance (Production - Consumption)")
plt.title("Hexbin Plot: Combined Relative Forecast Error vs Abs Energy Imbalance")
cb = plt.colorbar(hb)
cb.set_label("Count in Bin (log scale)")
plt.grid(True)
plt.show()

# -------------------------------
# Step 7: Compute the Correlation Coefficient
# -------------------------------
correlation = df_filtered["avg_combined_error"].corr(df_filtered["abs_imbalance"])
print("Correlation coefficient between average combined relative forecast error and energy imbalance:", correlation)
```

## Historical weather forecast correlation with energy imbalance -0.12

Below is the rewritten code that removes all forecast‐related computations and instead uses historical weather data. In this example, we create a “combined historical weather index” by computing z‑scores (how extreme the weather is) for several historical weather parameters (temperature, dewpoint, cloud cover, direct solar radiation, and snowfall) and averaging them. We then group by datetime, calculate the average combined historical weather index alongside the absolute energy imbalance (production minus consumption), filter out outliers, plot the relationship with a hexbin plot, and compute the correlation coefficient.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# Set plot style and parameters
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (19, 11)
plt.rcParams["font.size"] = 20
plt.rcParams["xtick.color"] = 'black'
plt.rcParams["ytick.color"] = 'black'
plt.rcParams["axes.edgecolor"] = 'black'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams["axes.labelcolor"] = 'black'
plt.rcParams["axes.linewidth"] = 1

# -------------------------------
# Step 1: Load the combined CSV file
# -------------------------------
df = pd.read_csv("train_features.csv",
                 parse_dates=['datetime', 'forecast_date_electricity', 'origin_date_electricity'])

# -------------------------------
# Step 2: Filter for Valid Historical Weather Data
# -------------------------------
# Define the historical weather parameters of interest.
params = ["temperature", "dewpoint", "cloudcover_high", "cloudcover_low",
          "cloudcover_mid", "cloudcover_total", "direct_solar_radiation", "snowfall"]
historical_cols = [f"{p}_h_mean" for p in params]

# Remove rows where any of the historical weather data is missing.
df = df.dropna(subset=historical_cols)

# -------------------------------
# Step 3: Compute Combined Historical Weather Index
# -------------------------------
# For each parameter, compute a z-score so that the different units/scales become comparable.
epsilon = 1e-5  # Small constant to avoid division by zero
for p in params:
    col = f"{p}_h_mean"
    mean_val = df[col].mean()
    std_val = df[col].std() + epsilon
    df[f"{p}_z"] = (df[col] - mean_val) / std_val

# Compute the combined historical weather index as the average of the z-scores.
# (You could also take a sum if preferred.)
df["combined_historical"] = df[[f"{p}_z" for p in params]].mean(axis=1)

# -------------------------------
# Step 4: Compute Energy Imbalance by Aggregating per Datetime
# -------------------------------
# We assume:
#   - Rows with is_consumption == 0 indicate production.
#   - Rows with is_consumption == 1 indicate consumption.
# Energy imbalance is defined as: Production - Consumption.
# For each datetime, compute:
#   - The average combined historical weather index.
#   - The absolute energy imbalance.
df_grouped = df.groupby("datetime", as_index=False).apply(lambda group: pd.Series({
    "avg_combined_historical": group["combined_historical"].mean(),
    "abs_imbalance": abs(
        group.loc[group["is_consumption"] == 0, "target"].sum() -
        group.loc[group["is_consumption"] == 1, "target"].sum()
    )
})).reset_index(drop=True)

# -------------------------------
# Step 5: Filter Out Outliers in Energy Imbalance
# -------------------------------
imbalance_lower = df_grouped["abs_imbalance"].quantile(0.01)
imbalance_upper = df_grouped["abs_imbalance"].quantile(0.99)
df_filtered = df_grouped[(df_grouped["abs_imbalance"] >= imbalance_lower) &
                         (df_grouped["abs_imbalance"] <= imbalance_upper)]

# -------------------------------
# Step 6: Graph the Correlation Using a Hexbin Plot
# -------------------------------
plt.figure(figsize=(8, 6))
hb = plt.hexbin(df_filtered["avg_combined_historical"], df_filtered["abs_imbalance"],
                gridsize=30, cmap='viridis', norm=LogNorm())
plt.xlabel("Average Combined Historical Weather Index (z-score)")
plt.ylabel("Absolute Energy Imbalance (Production - Consumption)")
plt.title("Hexbin Plot: Historical Weather vs Abs Energy Imbalance")
cb = plt.colorbar(hb)
cb.set_label("Count in Bin (log scale)")
plt.grid(True)
plt.show()

# -------------------------------
# Step 7: Compute the Correlation Coefficient
# -------------------------------
correlation = df_filtered["avg_combined_historical"].corr(df_filtered["abs_imbalance"])
print("Correlation coefficient between average combined historical weather and energy imbalance:", correlation)
```

# Ok what the heck has strong correlation with production then?? (Iterate through every historical datapoint, compare with production)

1.	Create separate columns for production and consumption by assigning target values based on is_consumption.
2.	For each time period, compute:
•	Total production (sum over rows where is_consumption is 0)
•	Total consumption (sum over rows where is_consumption is 1)
•	Energy imbalance = \text{abs}(\text{production} - \text{consumption})
3.	Also, aggregate the historical weather measurements. Since these columns are generally the same for both production and consumption entries at the same time point, taking their average (or “mean”) in the group suffices.
4.	Finally, iterate through every historical weather column (those with “_h_mean”) and compute its Pearson correlation with the energy imbalance.


```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("train_features.csv")

# Create separate production and consumption columns
# When is_consumption is 0, the target represents production; when 1, it represents consumption.
df['production'] = df.apply(lambda row: row['target'] if row['is_consumption'] == 0 else 0, axis=1)
df['consumption'] = df.apply(lambda row: row['target'] if row['is_consumption'] == 1 else 0, axis=1)

# Identify all historical weather columns (those containing "_h_mean")
historical_weather_cols = [col for col in df.columns if '_h_mean' in col]

# Define an aggregation dictionary:
# - Sum the production and consumption to get total values per time period.
# - Average the weather values (assuming they are recorded for the same timestamp, any minor differences are averaged).
aggregation_dict = {
    'production': 'sum',
    'consumption': 'sum'
}
for col in historical_weather_cols:
    aggregation_dict[col] = 'mean'

# Group the data by a time indicator, for example 'datetime'
grouped = df.groupby('datetime').agg(aggregation_dict).reset_index()

# Compute energy imbalance as the absolute difference between aggregated production and consumption.
grouped['imbalance'] = np.abs(grouped['production'] - grouped['consumption'])

# Iterate through each historical weather column and compute its correlation with energy imbalance.
correlations = {}
for col in historical_weather_cols:
    correlations[col] = grouped[col].corr(grouped['imbalance'])

# Report the correlation results.
print("Correlation between historical weather variables and energy imbalance:")
for col, corr_value in correlations.items():
    print(f"{col}: {corr_value}")
```

## Do forecasts strongly correlate with energy imbalance (i.e. what can we use to plan better?)

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("train_features.csv")

# Create production and consumption columns
df['production'] = df.apply(lambda row: row['target'] if row['is_consumption'] == 0 else 0, axis=1)
df['consumption'] = df.apply(lambda row: row['target'] if row['is_consumption'] == 1 else 0, axis=1)

# Identify all historical weather columns (those ending in '_h_mean')
historical_cols = [col for col in df.columns if col.endswith('_h_mean')]

# For each historical weather column, check if a corresponding forecast column exists.
# If yes, compute the forecast error (forecast - historical)
forecast_error_map = {}
for h_col in historical_cols:
    f_col = h_col.replace('_h_mean', '_f_mean')
    if f_col in df.columns:
        error_col = h_col.replace('_h_mean', '_error')
        # Compute forecast error for this weather variable
        df[error_col] = df[f_col] - df[h_col]
        forecast_error_map[h_col] = error_col

# Define aggregation rules:
# - Sum up production and consumption
# - Take the mean of forecast error columns (they should be similar within the same datetime)
agg_dict = {'production': 'sum', 'consumption': 'sum'}
for error_col in forecast_error_map.values():
    agg_dict[error_col] = 'mean'

# Group the DataFrame by datetime
grouped = df.groupby('datetime').agg(agg_dict).reset_index()

# Compute energy imbalance: absolute difference between total production and total consumption
grouped['imbalance'] = np.abs(grouped['production'] - grouped['consumption'])

# Now, compute the Pearson correlation between each forecast error and energy imbalance
correlations = {}
for error_col in forecast_error_map.values():
    correlations[error_col] = grouped[error_col].corr(grouped['imbalance'])

# Report the correlation results
print("Correlation between forecast error (forecast - historical) and energy imbalance:")
for error_col, corr_value in correlations.items():
    print(f"{error_col}: {corr_value}")
```
